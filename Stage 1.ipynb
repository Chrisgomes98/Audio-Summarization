```python
import audioread
import numpy as np

import matplotlib.pyplot as plt

audio=[]

with audioread.audio_open("./audio/good_morning.wav") as file:
    print(file.channels, file.samplerate, file.duration)
    freq_values = np.fft.fftfreq(file.samplerate, 1/file.samplerate)
    print(freq_values)
    for buffer in file:
        #print(buffer)
        audio.append(np.frombuffer(buffer,"Int16"))
```

```python
plt.title("Audio Wave Form")
plt.xlabel("Time in seconds")
plt.ylabel("Amplitude")

time=np.linspace(0,len(audio[:-1])/file.samplerate,num=len(audio[:-1]))*1000
plt.plot(time,audio[:-1],color="blue")

plt.rcParams["figure.figsize"] = (50,40)
plt.show()
#time vs amplitude
```

```python
from scipy import fft
transform = []

for i in audio:
    transform.append(fft(i))
    
time=np.linspace(0,len(transform[:-1])/file.samplerate,num=len(transform[:-1]))*1000
plt.plot(time,transform[:-1])
plt.rcParams["figure.figsize"] = (50,40)
plt.show()
```
